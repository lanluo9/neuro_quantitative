```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.animation as animation

import scipy.io
%matplotlib inline
```

```python
ori = scipy.io.loadmat('../data/week3/DirectionTuning_V1_dec.mat',struct_as_record=False)
ori.keys(), ori['data'].shape
```

```python
plt.imshow(ori['data'][:,:,0])
```

```python
fig, ax = plt.subplots()
ims = []
for i in range(ori['data'].shape[2] - 1):
    im = ax.imshow(ori['data'][:,:,i], animated=True)
    if i == 0:
        ax.imshow(ori['data'][:,:,0])
    ims.append([im])

ani = animation.ArtistAnimation(fig, ims, interval=50, blit=True, repeat_delay=1000)
plt.show()
```
